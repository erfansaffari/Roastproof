# Phase 2 — Corpus Exploration

Loads `data/structured/threads.jsonl`, produces charts under `notebooks/figs/`,
and writes `notebooks/FINDINGS.md`.

Prefer the CLI for a one-shot run:
```bash
python -m src.knowledge.norms
python -m src.knowledge.exploration
```

````python
from pathlib import Path
import json

from src.knowledge.norms import load_threads, compute_norms, run as run_norms
from src.knowledge.exploration import (
    threads_to_dataframe,
    label_critiques,
    generate_figures,
    write_findings,
)

THREADS = Path("data/structured/threads.jsonl")
FIGS = Path("notebooks/figs")
NORMS_JSON = Path("data/norms/norms.json")
NORMS_DB = Path("data/norms/norms.db")
FINDINGS = Path("notebooks/FINDINGS.md")

records = load_threads(THREADS)
print(f"Loaded {len(records)} threads")
df = threads_to_dataframe(records)
df.head()
````

````python
norms = run_norms(THREADS, NORMS_JSON, NORMS_DB)
norms["meta"]
````

````python
labeled = label_critiques(records, use_llm=False)
summary = generate_figures(df, labeled, records, FIGS)
write_findings(summary, norms, FINDINGS)
print("Figures:", summary["figures"])
print("Role distribution:", summary["role_distribution"])
FINDINGS.read_text()[:800]
````
